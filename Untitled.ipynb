```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('country_vaccinations_by_manufacturer.csv', sep=';')
df.head(15)
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df_2020 = df[df['date'].dt.year == 2020]
total_vacc_2020 = df_2020.groupby(df_2020['country'])['total_vaccinations'].sum().reset_index()

total_vacc_2020
```

```python
df_2021 = df[df['date'].dt.year == 2021]
total_vacc_2021 = df_2021.groupby(df_2021['country'])['total_vaccinations'].sum().reset_index()

total_vacc_2021
```

```python
df_2022 = df[df['date'].dt.year == 2022]
total_vacc_2022 = df_2022.groupby(df_2022['country'])['total_vaccinations'].sum().reset_index()

total_vacc_2022
```

```python
count_2020 = df_2020['vaccine'].value_counts().reset_index()
count_2020
```

```python
count_2021 = df_2021['vaccine'].value_counts().reset_index()
count_2021
```
